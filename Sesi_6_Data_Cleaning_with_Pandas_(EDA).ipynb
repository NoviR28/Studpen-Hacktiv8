**Non-Standart Missing Values**

  merupakan missing value dimana nilainya selain Nan/NA. seperti:


*  na, n/a, NA, NAN
*  ---, ..., dst


Cara menetapkan missing values pada suatu dataframe dapat dilakukan menguunakan parameter pandas yaitu na_values = .... .



```python
# Untuk mempermudah mengecek missing values. Kita kumpulkan terlebih dahulu nilai dari missing values tersebut.

miss_values =["n/a", "na", "--", "HURLEY"]
```

```python
# Menetapkan missing values pada
data1= pd.read_csv(path_property, na_values =miss_values)

data1.head(10)
```

```python
data1.isnull().head(10)

# Sekarang, setiap nilai yang termasuk ke dalam missing values akan otomatis terdeteksi sebagai missing values oleh pandas
```

**Unexpected Missing Values**

  merupakan missing values yang diakibatkan oleh nilai yang tidak terduga seperti pada kolom own_occupied dan num_bath dimana muncul angka 12 dan HURLEY yang berbeda tipe dengan nilai lainnya. Oleh karena itu, pengecekkan missing value bisa dilakukan menggunakan

  **try ................... except ValueError**

**Kolom OWN_OCCUPIED**

```python
# Cek pada kolom own_occupied

count = 0
for row in data1["OWN_OCCUPIED"]:
  try:
    int(row)
    data1.loc[count,"OWN_OCCUPIED"] =np.nan   # np.nan dirubah menjadi missing value
  except ValueError:
    pass
    count+=1

data1

# Sekarang missing value pada kolom OWN_OCCUPIED sudah dirubah dalam bentuk NAN
```

```python
# Merubah tipe data pada kolom "OWN_OCCUPIED" ke dalam categorical

type= pd.CategoricalDtype(categories=['N', 'Y'], ordered=True)

data1["OWN_OCCUPIED"]= data1["OWN_OCCUPIED"].astype(type)

data1.info()
```

```python
# Cari modus dari kolom "OWN_OCCUPIED"
modus = data1["OWN_OCCUPIED"].mode()
modus

# Berdasarkan hasil di atas dapat disimpulkan bahwa modus adalah "Y". Sehingga, missing value akan diisi dengan "Y"
```

```python
# Mereplace missing value dengan modus

data1["OWN_OCCUPIED"].fillna("Y", inplace= True)

data1
```

**KOLOM NUM_BEDROOMS**

```python
# Mencari mean untuk mengisi missing value

data1["NUM_BEDROOMS"].mean()

# hasilnya 2.4
```

```python
# Masukkan mean ke dalam kolom

data1["NUM_BEDROOMS"].fillna(2.4, inplace=True)
data1
```

```python
# Ubah tipe data " NUM_BEDROOMS" ke dalam int

data1["NUM_BEDROOMS"] = data1["NUM_BEDROOMS"].astype("int")
data1.info()
```
