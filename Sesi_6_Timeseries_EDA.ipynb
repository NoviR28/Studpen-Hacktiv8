```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
path_opsd = 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv'

data= pd.read_csv(path_opsd)

data
```

**Membuat range waktu dengan datetime**

```python
waktu =pd.date_range(start='1/1/2006',end= '31/12/2017', freq='H')
waktu = waktu.normalize()   # normalize() untuk mensetting waktu pada pukul 00:00 sehingga yang ditampilkan hanya tanggal/bulan/tahun saja.
waktu
```

### 1. Memasukkan data di atas ke dalam dataframe karena **jika ingin memiliki data timeseries** maka kita **memerlukan time index**.

**Membuat kolom date dengan tipe datetime**. Lalu **memasukkannya ke dalam Dataframe**.

```python
data["date"] = pd.DataFrame(waktu, columns=["date"])

data.head(10)
```
