### Melakukan **Transpose**



Metode **squeeze()** pada pandas digunakan untuk mengubah DataFrame dengan satu kolom menjadi Series. Jika DataFrame hanya memiliki satu kolom, maka metode ini akan menghasilkan Series. Jika DataFrame memiliki lebih dari satu kolom atau lebih dari satu baris, maka objek tersebut tidak akan berubah



```python
data[data.index=="Japan"].T.squeeze()
```

```python
# Contoh slicing menggunakan loc


data.loc["Japan",[2010,2011,2012,2013]]
data.iloc[87, 3:9]
```

### **Merubah tipe judul kolom** menjadi **string**


Rumus = **df_can.columns = list(map(str, df_can.columns))**

- Fungsi **df_can.columns** digunakan untuk mengambil nama kolom dari DataFrame df_can.  
- Fungsi **map(str)** digunakan untuk mengubah tipe data setiap judul kolom menjadi string.
- Fungsi **list()** digunakan untuk mengubah hasil dari fungsi map() menjadi list.

```python
data.columns = list(map(str, data.columns))

data.info()
```

Karena tipe judul kolom telah berubah menjadi string. Maka perlu **mendeklarasikan variabel berisi rentang waktu** yang bisa digunakan untuk melakukan visualisasi, dll.

```python
# Variabel rentang waktu yang bisa sewaktu-waktu dipanggil
years= list(map(str, range(1980, 2014)))

years
```
