**Exploring Datasets**

- Melihat seberapa sering suatu nilai muncul dalam data menggunakan .value_counts()  => namatabel["Label"]..value_counts()

```python
data["team_id"].value_counts()
```

```python
data["fran_id"].value_counts()
```

Tim "Lakers" memainkan sebanyak 6024 pertandingan. Tetapi hanya 5078 yang dimainkan oleh Los Angeles Lakers(LAL). Selanjutnya mencari tahu siapa yang memanikan selisih tersebut.

```python
data.loc[data["fran_id"] == "Lakers", "team_id"].value_counts()
```

Berdasarkan hasil di atas, diketahui bahwa
Tim lakers yang memainkan pertandingan yaitu:
  - Los Angeles Lakers(LAL) sebanyak 5078.
  - Minneapolis Lakers (MNL) sebanyak 97 pertandingan




  Lalu, jika Kita juga bisa mengetahui waktu MNL memainkan pertandingan tersebut.

```python
# Pertandingan paling awal yang dilakukan

data.loc[data["team_id"]=="MNL", "date_game"].min()

# Pertandingan paling akhir yang dilakukan

data.loc[data["team_id"]=="MNL", "date_game"].max()


# Pertandingan paling awal dan akhir

data.loc[data["team_id"]=="MNL", "date_game"].agg(("min","max"))
```

Membedah alasan tim **BOS** (Boston Celtics) memainkan game paling banyak


- Tim **BOS** paling sering menang

```python
# Paling sering menang

data.loc[data["game_result"] == "W", "team_id"].value_counts()
```
